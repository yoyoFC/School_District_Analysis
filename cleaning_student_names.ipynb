```python
import pandas as pd
import os

#Load external files
student_data_load = os.path.join("Resources", "students_complete.csv")
student_data_df = pd.read_csv(student_data_load)
student_data_df.head()
```

```python
#Put the student name in a list - tolist()
students_to_fix = []
student_names = student_data_df["student_name"].tolist()
#Split the names and determine the lenght of names
for name in student_names:
    if len(name.split()) >=3:
        students_to_fix.append(name)
        
#Get the number of student with 3 or more names
len(students_to_fix)
    
```

```python
#Additional example to play with split
#name = "Mrs. Linda Fernandez"
#name.split()
```

```python
#Get All the prefixes
prefixes = []
for name in students_to_fix:
    if len(name.split()[0]) <= 4:
       prefixes.append(name.split()[0])
prefixes
```

```python
# Get the unique items in the "prefixes" list - set
set(prefixes)
```

```python
#Get All the suffixes
suffixes = []
for name in students_to_fix:
    if len(name.split()[-1]) <= 3:
       suffixes.append(name.split()[-1])
suffixes
```

```python
# Get the unique items in the "suffixes" list.
set(suffixes)
```

```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]

#Keep in mind the replace method only works in string
#lets convert the data type of the name column as string
#before start analyzing inside of the for room 

for word in prefixes_suffixes:
        student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

student_data_df.head()
```

```python
#Confirm that all the replacement procedure was correct
new_student_name = student_data_df["student_name"].tolist()
print(new_student_name)

# Create a new list and use it for the for loop to iterate through the list.
students_fixed = []
for name in new_student_name:
    if len(name.split()) >= 3:
        student_fixed.append(name)
        
# Get the length of the students' names that are greater than or equal to 3.
len(students_fixed)
```
